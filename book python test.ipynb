```python
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\huysi\Desktop\book_collection.csv")
df = pd.DataFrame(data)

ax = df['reading_status'].value_counts().plot(kind='bar', title="Huy's Reading Status")

# Set labels
plt.xlabel('Status')
plt.ylabel('Book Count')

# Set y-axis to show only integers and limit range to 0–20
plt.ylim(0, 16)
ax.set_yticks(range(0, 16, 1))  # Tick marks at every integer up to 20

plt.show()
```
